## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
cur.execute('select distinct(Launch_site) from SPACEXTBL')
print(cur.fetchall())
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
cur.execute("select * from SPACEXTBL WHERE launch_site like 'CCA%'")
for rec in cur.fetchall():
    print(rec)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
cur.execute("select SUM(PAYLOAD_MASS__KG_) from SPACEXTBL where customer = 'NASA (CRS)'")
print(cur.fetchall())
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
cur.execute("select AVG(PAYLOAD_MASS__KG_) from SPACEXTBL where Booster_Version = 'F9 v1.1'")
print(cur.fetchall())
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
cur.execute("select MIN(DATE) from SPACEXTBL where LANDING_OUTCOME = 'Success (ground pad)'")
print(cur.fetchall())
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
cur.execute("select BOOSTER_VERSION from SPACEXTBL where LANDING_OUTCOME = 'Success (drone ship)' AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000")
print(cur.fetchall())
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
cur.execute("select MISSION_OUTCOME,COUNT(1) from SPACEXTBL WHERE MISSION_OUTCOME != 'None' GROUP BY MISSION_OUTCOME")
print(cur.fetchall())
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
cur.execute("select distinct booster_version from SPACEXTBL WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL)")
print(cur.fetchall())
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
cur.execute("select substr(date,4,2),landing_outcome,booster_version,launch_site from SPACEXTBL WHERE landing_outcome = 'Failure (drone ship)' and substr(Date,7,4) = '2015'")
print(cur.fetchall())
```

### Task 10




##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
cur.execute("select COUNT(1),landing_outcome from SPACEXTBL WHERE landing_outcome LIKE 'Success%' and DATE BETWEEN '04-06-2010' AND '20-03-2017'" 
           +" group by landing_outcome order by count(1) desc")
print(cur.fetchall())
```
